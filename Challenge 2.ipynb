```python
import numpy as np

def angle_between(p1, p2):
    ang1 = np.arctan2(*p1[::-1])
    ang2 = np.arctan2(*p2[::-1])
    return np.rad2deg((ang1 - ang2))

A = [-1.0, -0.02]
B = [-1.0, 0.02]
angle_between(A, B)
```

```python
import csv
with open('circle.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    point_id = []
    point_x = []
    point_y = []
    for row in readCSV:
        id_ = row[0]
        x = row[1]
        y = row[2]

        point_id.append(id_)
        point_x.append(x)
        point_y.append(y)

point_id = map(int, point_id[1:])
point_x = map(float, point_x[1:])
point_y = map(float, point_y[1:])

print len(point_id)
print point_id
```

```python
import numpy as np

angles = np.arctan2(point_y, point_x)
print len(angles)
```

```python
angle_between = [np.rad2deg(ang1 - ang2) for ang1, ang2 in zip(angles, angles[1:])]
len(angle_between)
print angle_between
```

```python
for a in angle_between:
    if (a < 0 and a > -180):
        print point_id[angle_between.index(a) + 1]
```
